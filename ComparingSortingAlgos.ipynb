## Insertion Sort

```python
def insertionSort(arr): 
  
    # Traverse through 1 to len(arr) 
    for i in range(1, len(arr)): 
        key = arr[i] 
        # Move elements of arr[0..i-1], that are greater than key, to one position ahead of their current position 
        j = i-1
        while j >= 0 and key < arr[j] : 
                arr[j + 1] = arr[j] 
                j -= 1
        arr[j + 1] = key 
        

print("\033[1m" + "                    INSERTION SORT                  " + "\033[0m")
print("----------------------------------------------------")

numList = [33,21,4,90,3,2,10,22,74,92]

n = len(numList) 
print("List before sorting : \n")
for i in range(n): 
    print ("%d" %numList[i])
    
    
    
insertionSort(numList) 
print("\nList after sorting : \n")
for i in range(n): 
    print ("%d" %numList[i])
```

## Megre Sort ##

```python
def merge(arr, l, m, r): 
    n1 = m - l + 1
    n2 = r- m 
  
    # create temp arrays 
    L = [0] * (n1) 
    R = [0] * (n2) 
  
    # Copy data to temp arrays L[] and R[] 
    for i in range(0 , n1): 
        L[i] = arr[l + i] 
  
    for j in range(0 , n2): 
        R[j] = arr[m + 1 + j] 
  
    # Merge the temp arrays back into arr[l..r] 
    i = 0     # Initial index of first subarray 
    j = 0     # Initial index of second subarray 
    k = l     # Initial index of merged subarray 
  
    while i < n1 and j < n2 : 
        if L[i] <= R[j]: 
            arr[k] = L[i] 
            i += 1
        else: 
            arr[k] = R[j] 
            j += 1
        k += 1
  
    # Copy the remaining elements of L[], if there are any 
    while i < n1: 
        arr[k] = L[i] 
        i += 1
        k += 1
  
    # Copy the remaining elements of R[], if there are any 
    while j < n2: 
        arr[k] = R[j] 
        j += 1
        k += 1
        
        
        
        
# l is for left index and r is right index of the sub-array of arr to be sorted 
def mergeSort(arr,l,r): 
    if l < r: 
  
        # Same as (l+r)//2, but avoids overflow for large l and h 
        m = (l+(r-1))//2
  
        # Sort first and second halves 
        mergeSort(arr, l, m) 
        mergeSort(arr, m+1, r) 
        merge(arr, l, m, r) 
  
  
print("\033[1m" + "                    MERGE SORT                  " + "\033[0m")
print("------------------------------------------------")

arr = [33,21,4,90,3,2,10,22,74,92]
n = len(arr) 
print("\n\n\nList before sorting : ")
for i in range(n): 
    print ("%d" %arr[i])
    
    
    
mergeSort(arr,0,n-1) 
print("\nList after sorting : ")
for i in range(n): 
    print ("%d" %arr[i])
```

## Bubble Sort ##

```python
def bubble_sort(array):
    index = len(array) - 1
    while index >= 0:
        for j in range(index):
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j]
        index -= 1
    return array



print("\033[1m" + "                    BUBBLE SORT                  " + "\033[0m")
print("-------------------------------------------------")
num = [54, 26, 93, 17, 77, 31, 44, 55, 20]
n = len(num) 
print("\n\nList before sorting : \n")
for i in range(n): 
    print ("%d" %num[i])
    
    
    
bubble_sort(num) 
print("\n\nList after sorting : \n")
for i in range(n): 
    print ("%d" %num[i])
```

## Selection Sort ##

```python
def SelectionSort(numList):
    for i in range(len(numList)):
        min_index = i
        for j in range(i+1, len(numList)):
            if numList[min_index] > numList[j]:
                min_index = j
        
        #swapping the element
        numList[i], numList[min_index] = numList[min_index], numList[i]
        
        
numList = [54, 26, 93, 17, 77, 31, 44, 55, 20]
print("\033[1m" + "                    SELECTION SORT                  " + "\033[0m")
print("----------------------------------------------------")
print("\n\nList before sorting : \n")
for i in numList:
    print(i)

    
print("\n\n")
print("List after sorting : \n")
SelectionSort(numList)
for i in numList:
    print(i)
```

## Quick Sort ##

```python
def pivot(arr, low, high):
    i = (low - 1)
    myPivot = arr[high]
    for x in range(low, high):
        if arr[x] <= myPivot:
            i+=1
            arr[i], arr[x] = arr[x], arr[i]
            
    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    
    return i+1


def QuickSort(arr, low, high):
    if len(arr) == 1:
        return arr
    if low < high:
        pi = pivot(arr, low, high)
        QuickSort(arr, low, pi-1)
        QuickSort(arr, pi+1, high)
        
        
        
print("\033[1m" + "                    QUICK SORT                  " + "\033[0m")
print("-------------------------------------------------")
num = [54, 26, 93, 17, 77, 31, 44, 55, 20]
n = len(num) 
print("\n\nList before sorting : \n")
for i in range(n): 
    print ("%d" %num[i])
    
    
    
QuickSort(num, 0, n-1) 
print("\n\nList after sorting : \n")
for i in range(n): 
    print ("%d" %num[i])
```

## Shell Sort ##

```python
def ShellSort(alist):
    n = len(alist)//2
    while n > 0:
        for i in range(n):
            gap_InsertionSort(alist, i, n)
            
        n = n//2
        
def gap_InsertionSort(alist, start, gap):
    for i in range(start + gap, len(alist), gap):
        cur_val = alist[i]
        pos = i
        
        while pos >=gap and alist[pos - gap] > cur_val:
            alist[pos] = alist[pos - gap]
            pos = pos - gap
            
        alist[pos] = cur_val

        
        
        
numList = [54, 26, 93, 17, 77, 31, 44, 55, 20]
print("\033[1m" + "                    SHELL SORT                  " + "\033[0m")
print("----------------------------------------------------")
print("\n\nList before sorting : \n")
for i in numList:
    print(i)

    
print("\n\n")
print("List after sorting : \n")
ShellSort(numList)
for i in numList:
    print(i)
```

```python
import timeit
print("Insertion Sort Time : ")
%timeit insertionSort(numList)

n=len(numList)
print("Merge Sort Time : ")
%timeit mergeSort(numList, 0, n-1)

print("Bubble Sort Time : ")
%timeit bubble_sort(numList)

print("Selection Sort Time : ")
%timeit SelectionSort(numList)

print("Quick Sort Time : ")
%timeit QuickSort(numList, 0, n-1)

print("Shell Sort Time : ")
%timeit ShellSort(numList)
```
